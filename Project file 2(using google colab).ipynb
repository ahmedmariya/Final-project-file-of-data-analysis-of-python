```python
import pandas as pd
import matplotlib.pyplot as plt
path = "/content/Dataset.csv"
data = pd.read_csv(path)

# Line chart: Comparison of expected accuracies for the first 15 data points
plt.figure(figsize=(10, 6))
plt.plot(data['focal_depth'][:15], label='Focal depth', marker='o')
plt.title('Expected Accuracy Comparison (First 10 Data Points)')
plt.xlabel('Data Point Index')
plt.ylabel('Expected Accuracy')
plt.legend()
plt.grid(True)
plt.show()

# Pie chart: Distribution of the first 15 'latitude' categories
latitude_counts = data['latitude'][:15].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(latitude_counts, labels=latitude_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('latitude Distribution (First 5 Data Points)')
plt.show()

# Histogram: Distribution of 'longitude'
plt.figure(figsize=(8, 6))
plt.hist(data['longitude'], bins=15, color='blue', edgecolor='black')
plt.title('Distribution of longitude')
plt.xlabel('longitude')
plt.ylabel('longitude Frequency')
#plt.grid(True)
plt.show()

# Scatter plot: Comparison between 'richter' and 'focal_depth'
plt.figure(figsize=(8, 6))
plt.scatter(data['richter'], data['focal_depth'], alpha=0.7, color='red')
plt.title('Scatter Plot: Comparison between richter and focal_depth')
plt.xlabel('Richter')
plt.ylabel('focal_depth')
#plt.grid(True)
plt.show()
```
